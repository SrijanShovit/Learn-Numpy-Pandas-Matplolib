```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(11)
y = x ** 3
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("t")
plt.xlabel("t^3")
plt.title("Cubic graph")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```python
#object oriented approach
fig = plt.figure() #blank canvas
axes = plt.axes([0,0,1,1]) #left,bottom,canvas' width,canvas' height
axes.plot(x,y)
axes.set_xlabel('my x label')
axes.set_ylabel('my y label')
axes.set_title('my title')
```

```python
fig = plt.figure(figsize=(10,2))
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(x,y)
axes1.set_title('Bigger')
axes2.set_title('Smaller')
```

```python
fig,axes = plt.subplots(3,2)
```

```python
#object oriented approach for subplots
fig,axes = plt.subplots(1,2)
#alpha for opacity
axes[0].plot(x,y,color='#00ff7b',linewidth=5,alpha=0.4)
axes[1].plot(x,y,'g',linestyle='-.',marker='o',markerfacecolor='yellow',markersize='15',markeredgecolor='r')
```

```python
fig,axes = plt.subplots(1,2)
for ax in axes:
    ax.plot(x,y)
plt.show()
```

```python
#saving chart
fig.savefig('mychart.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label="cubic")
axes.plot(x,x**2,label="quadratic")
axes.legend(loc=1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.bar(x,y,label="cubic")
axes.bar(x,x**2,label="quadratic")
axes.legend(loc=1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.bar(x,y,label="cubic",width=0.4)
axes.bar(x+0.4,x**2,label="quadratic",width=0.4)
plt.xticks(x+0.2,['Ashish','Bumrah','Cheteshwar','Dinesh','Eshwar','Fuhar','Gabriel','Hrish','Isha','John','Karim'],rotation=60)
axes.legend()
```
